```python
def calculate_adj_close_range_ratio(ticker_symbol, start_date, end_date):
  data = yf.download(ticker_symbol, start=start_date, end=end_date)
  min = data['Adj Close'].min()
  max = data['Adj Close'].max()
  diff = max - min
  return diff / max
```

```python
start_date = '2023-01-01'
end_date = '2023-12-31'
sr_ratio = calculate_adj_close_range_ratio('2222.SR', start_date, end_date)
brk_ratio = calculate_adj_close_range_ratio('BRK-B', start_date, end_date)
aapl_ratio = calculate_adj_close_range_ratio('AAPL', start_date, end_date)
msft_ratio = calculate_adj_close_range_ratio('MSFT', start_date, end_date)
goog_ratio = calculate_adj_close_range_ratio('GOOG', start_date, end_date)
jpm_ratio = calculate_adj_close_range_ratio('JPM', start_date, end_date)
```

```python
print(f"sr_ratio: {sr_ratio}")
print(f"brk_ratio: {brk_ratio}")
print(f"aapl_ratio: {aapl_ratio}")
print(f"msft_ratio: {msft_ratio}")
print(f"goog_ratio: {goog_ratio}")
print(f"jpm_ratio: {jpm_ratio}")
```

```python
def calculate_dividend_yield(ticker_symbol, start_date, end_date):
  df = yf.download(ticker_symbol, start=start_date, end=end_date)
  dividends = yf.Ticker(ticker_symbol).dividends
  print(f"\nticker_symbol: {ticker_symbol}")
  print(f"\ndividends: {dividends}")
  total_dividends = dividends[str(2023)].sum()
  print(f"\ntotal_dividends: {total_dividends}")
  last_closing_price = df.iloc[-1]['Adj Close']
  print(f"\nlast_closing_price: {last_closing_price}")
  return total_dividends / last_closing_price
```

```python
SR_SYMBOL = '2222.SR'
BRK_RATIO = 'BRK-B'
AAPL_RATIO = 'AAPL'
MSFT_RATIO = 'MSFT'
GOOG_RATIO = 'GOOG'
JPM_RATIO = 'JPM'
```

```python
start_date = '2023-01-01'
end_date = '2023-12-31'

sr_dividend_yield = calculate_dividend_yield(SR_SYMBOL, start_date, end_date)
brk_dividend_yield = calculate_dividend_yield(BRK_RATIO, start_date, end_date)
aapl_dividend_yield = calculate_dividend_yield(AAPL_RATIO, start_date, end_date)
msft_dividend_yield = calculate_dividend_yield(MSFT_RATIO, start_date, end_date)
goog_dividend_yield = calculate_dividend_yield(GOOG_RATIO, start_date, end_date)
jpm_dividend_yield = calculate_dividend_yield(JPM_RATIO, start_date, end_date)
```

```python
print(f"sr_dividend_yield: {sr_dividend_yield}")
print(f"brk_dividend_yield: {brk_dividend_yield}")
print(f"aapl_dividend_yield: {aapl_dividend_yield}")
print(f"msft_dividend_yield: {msft_dividend_yield}")
print(f"goog_dividend_yield: {goog_dividend_yield}")
print(f"jpm_dividend_yield: {jpm_dividend_yield}")
```
